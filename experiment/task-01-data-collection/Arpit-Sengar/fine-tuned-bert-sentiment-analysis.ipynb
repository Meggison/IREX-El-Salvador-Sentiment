```python
import pandas as pd
df=messages = pd.read_csv('data/twitter_dataset.csv', sep=',', names=["Tweet ID", "entity", "sentiment", "Tweet content"])
df.head()
```

```python
df = df[df['sentiment'] != 'Irrelevant']
```

```python
df.shape
```

```python
X=list(df['Tweet content'])
```

```python
y=list(df['sentiment'])
```

```python
y
```

```python
y=list(pd.get_dummies(y,drop_first=True)['Positive'])

y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
X
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
y_train
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
from transformers import TFDistilBertForSequenceClassification, TFTrainingArguments, TFTrainer

training_args = TFTrainingArguments(
    output_dir='./results',   
    num_train_epochs=3,              
    per_device_train_batch_size=8,  
    per_device_eval_batch_size=8,    
    warmup_steps=500,                
    weight_decay=0.01,               
    logging_dir='./logs',            
    logging_steps=10,                
    evaluation_strategy="steps",     
    eval_steps=500,                 
)
```

```python
with training_args.strategy.scope():
    model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = TFTrainer(
    model=model,                        
    args=training_args,                 
    train_dataset=train_dataset,         
    eval_dataset=test_dataset         
)

trainer.train()
```

```python
trainer.evaluate(test_dataset)
```

```python
trainer.predict(test_dataset)
```

```python
trainer.predict(test_dataset)[1].shape
```

```python
output=trainer.predict(test_dataset)[1]
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,output)
cm
```

```python
trainer.save_model('senti_model')
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,output)
acc
```
